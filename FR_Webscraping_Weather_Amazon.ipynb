# Information about the local weather from the National Weather Service website of USA

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.7771&lon=-122.4196#.YAPQ_ugzbDc')
```

```python
page.content
```

```python
soup= BeautifulSoup(page.content,'html.parser')
```

```python
period=[]
Sdescription=[]
temperature=[]
description=[]
```

```python
prd=soup.find_all('p',class_='period-name')
prd
```

```python
for i in prd:
    period.append(i.text)
    
period=period[0:7]
```

```python
Sdescr=soup.find_all('p',class_='short-desc')
Sdescr
```

```python
for i in Sdescr:
    Sdescription.append(i.text)

Sdescription=Sdescription[0:7]
```

```python
temp=soup.find_all('p',class_='temp')
temp
```

```python
for i in temp:
    temperature.append(i.text)
    
temperature=temperature[0:7]
temperature
```

```python
dscr=soup.find_all('div',class_='col-sm-10 forecast-text')
dscr
```

```python
for i in dscr:
    description.append(i.text)
    
description=description[0:7]
```

```python
print(len(temperature),len(period),len(Sdescription),len(description))
```

```python
weather=pd.DataFrame({})
weather['Deriods']=period
weather['Short_Descriptions']=Sdescription
weather['Temperature']=temperature
weather['Descriptions']=description

weather
```
