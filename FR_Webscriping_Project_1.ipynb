# Men's ODI Team Rankings

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page= requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

```python
page.content
```

```python
soup= BeautifulSoup(page.content,'html.parser')
```

```python
Team=[]
Matches=[]
Points=[]
Rating=[]
```

```python
Country = soup.find_all('span',class_="u-hide-phablet")
Country[0:4]
```

```python
for i in Country:
    Team.append(i.get_text().replace("\n",""))
    
Team=Team[0:10]
Team
```

```python
match=soup.find_all('td',class_='rankings-block__banner--matches')

matchs=soup.find_all('td',class_='table-body__cell u-center-text')

mtc = match + matchs

for i in mtc:
    Matches.append(i.text)
    Matches=Matches[0:10]
```

```python
Matches
```

```python
pt=soup.find_all('td',class_="rankings-block__banner--points")

pts= soup.find_all('td',class_ ="table-body__cell u-center-text")
Point= pt + pts
```

```python
for i in Point:
    Points.append(i.get_text().replace("\n",""))
    Points=Points[0:10]
```

```python
rating = soup.find_all('td',class_="table-body__cell u-text-right rating")
rating[0:4]
```

```python
for i in rating:
    Rating.append(i.get_text().replace("\n",""))
    Rating=Rating[0:10]
```

```python
print(len(Team),len(Rating),len(Points))
```

```python
ODI=pd.DataFrame({})
ODI['Country']=Team
ODI['Matches']=Matches
ODI['Rating']=Rating
ODI['Points']=Points
```

```python
ODI
```

# Information about the local weather from the National Weather Service website of USA

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page3=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.7771&lon=-122.4196#.YAPQ_ugzbDc')
```

```python
page3.content
```

```python
soup3= BeautifulSoup(page3.content,'html.parser')
```

```python
period=[]
Sdescription=[]
temperature=[]
description=[]
```

```python
prd=soup3.find_all('p',class_='period-name')
prd
```

```python
for i in prd:
    period.append(i.text)
    
period=period[0:7]
```

```python
Sdescr=soup3.find_all('p',class_='short-desc')
Sdescr
```

```python
for i in Sdescr:
    Sdescription.append(i.text)

Sdescription=Sdescription[0:7]
```

```python
temp=soup3.find_all('p',class_='temp')
temp
```

```python
for i in temp:
    temperature.append(i.text)
    
temperature=temperature[0:7]
temperature
```

```python
dscr=soup3.find_all('div',class_='col-sm-10 forecast-text')
dscr
```

```python
for i in dscr:
    description.append(i.text)
    
description=description[0:7]
```

```python
print(len(temperature),len(period),len(Sdescription),len(description))
```

```python
weather=pd.DataFrame({})
weather['Deriods']=period
weather['Short_Descriptions']=Sdescription
weather['Temperature']=temperature
weather['Descriptions']=description

weather
```
